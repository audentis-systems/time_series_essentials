```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from time_series_essentials.time_conversions import seconds_in_one_day
from time_series_essentials.time_conversions import seconds_in_one_week

from time_series_essentials.day_and_week.day_and_week import day_sin, day_cos
from time_series_essentials.day_and_week.day_and_week import week_sin, week_cos
```

```python
t_day = np.arange(0, 2. * seconds_in_one_day, 60)
t_week = np.arange(0, 2. * seconds_in_one_week, 60)
```

```python
y_day_sin = day_sin(t_day)
y_day_cos = day_cos(t_day)
y_week_sin = week_sin(t_week)
y_week_cos = week_cos(t_week)
```

```python
plt.figure()
plt.plot(t_day / seconds_in_one_day, y_day_sin)
plt.plot(t_day / seconds_in_one_day, y_day_cos)
plt.title('Two Days')
plt.xlabel('Day')
plt.ylabel('Amplitude')
plt.show()
plt.close()
```

```python
plt.figure()
plt.plot(t_week / seconds_in_one_day, y_week_sin)
plt.plot(t_week / seconds_in_one_day, y_week_cos)
plt.title('Two Weeks')
plt.xlabel('Day')
plt.ylabel('Amplitude')
plt.show()
plt.close()
```
